```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de X y sus probabilidades
x = np.array([0, 1, 2, 3, 4, 5])
p = np.array([0.10, 0.25, 0.30, 0.20, 0.10, 0.05])

# Verificar que las probabilidades sumen 1
suma_prob = np.sum(p)
print("Suma de probabilidades:", suma_prob)

# Distribución de probabilidad (ya definida)
print("\nDistribución de probabilidad:")
for xi, pi in zip(x, p):
    print(f"P(X={xi}) = {pi}")

# Cálculos
# Valor esperado
E_X = np.sum(x * p)
print("\nValor esperado E(X):", E_X)

# Varianza
var_X = np.sum((x - E_X)**2 * p)
print("Varianza Var(X):", var_X)

# Desviación estándar
std_X = np.sqrt(var_X)
print("Desviación estándar:", std_X)

# Función de distribución acumulada (CDF)
F = np.cumsum(p)

print("\nFunción de distribución acumulada F(x):")
for xi, Fi in zip(x, F):
    print(f"F({xi}) = {Fi}")

# Gráfica de la CDF
plt.figure(figsize=(8,5))
plt.step(x, F, where='post')
plt.scatter(x, F)
plt.title("Función de Distribución Acumulada (CDF)")
plt.xlabel("Número de células anormales (X)")
plt.ylabel("F(x)")
plt.grid()
plt.show()
```
