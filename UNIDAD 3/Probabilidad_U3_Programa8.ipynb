```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Definir la distribución
data = {
    'X (unidades)': [0, 1, 2, 3, 4, 5],
    'P(X)': [0.05, 0.10, 0.25, 0.30, 0.20, 0.10]
}

df = pd.DataFrame(data)
print("Distribución:")
print(df.to_string(index=False))

# Verificar suma de probabilidades
print("\nSuma de probabilidades:", df['P(X)'].sum())

# Valor esperado
E_X = np.sum(df['X (unidades)'] * df['P(X)'])
print("\nValor esperado E(X):", E_X)

# Varianza y desviación estándar
varianza = np.sum(df['P(X)'] * (df['X (unidades)'] - E_X)**2)
desv_std = np.sqrt(varianza)

print("Varianza:", varianza)
print("Desviación estándar:", desv_std)

# Función de distribución acumulada (CDF)
df = df.sort_values(by='X (unidades)').reset_index(drop=True)
df['F(x)'] = df['P(X)'].cumsum()

print("\nDistribución acumulada:")
print(df.to_string(index=False))

# Gráfica de la CDF
plt.figure()
plt.step(df['X (unidades)'], df['F(x)'], where='post')
plt.xlabel('X (unidades)')
plt.ylabel('F(x)')
plt.title('Función de Distribución Acumulada')
plt.grid()
plt.show()
```
