```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 1. Definir la distribución
# -----------------------------
data = {
    'Premio': ['A', 'B', 'C', 'D'],
    'X (pesos)': [0, 10, 20, 50],
    'P(X)': [0.50, 0.30, 0.15, 0.05]
}

df = pd.DataFrame(data)
print("Distribución:")
print(df)

# -----------------------------
# 2. Verificar que las probabilidades sumen 1
# -----------------------------
suma_prob = df['P(X)'].sum()
print("\nSuma de probabilidades:", suma_prob)

# -----------------------------
# 3. Valor esperado E(X)
# -----------------------------
E_X = np.sum(df['X (pesos)'] * df['P(X)'])
print("\nValor esperado E(X):", E_X)

# -----------------------------
# 4. Varianza y desviación estándar
# -----------------------------
varianza = np.sum(df['P(X)'] * (df['X (pesos)'] - E_X)**2)
desv_std = np.sqrt(varianza)

print("Varianza:", varianza)
print("Desviación estándar:", desv_std)

# -----------------------------
# 5. Función de distribución acumulada (CDF)
# -----------------------------
df = df.sort_values(by='X (pesos)')
df['F(x)'] = df['P(X)'].cumsum()

print("\nDistribución acumulada:")
print(df)

# -----------------------------
# 6. Gráfica de la CDF
# -----------------------------
plt.figure()
plt.step(df['X (pesos)'], df['F(x)'], where='post')
plt.xlabel('X (pesos)')
plt.ylabel('F(x)')
plt.title('Función de Distribución Acumulada')
plt.grid()

plt.show()
```
