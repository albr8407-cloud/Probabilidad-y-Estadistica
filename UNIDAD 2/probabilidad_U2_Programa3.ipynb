## Programa 3: Permutaciones

```python
import math

# Datos del problema
n = 10  # tamaño del grupo (asistentes)
r = 4   # tamaño del subconjunto (preguntas)

# Cálculo de la permutación
permutaciones = math.factorial(n) / math.factorial(n - r)

# Mostrar resultados paso a paso
print("Ejercicio de Permutaciones")
print("--------------------------")
print("Número total de asistentes (n):", n)
print("Número de preguntas (r):", r)

print("\nFórmula usada:")
print("P(n,r) = n! / (n-r)!")

print("\nSustituyendo valores:")
print("P(10,4) = 10! / (10-4)!")

print("\nCálculo:")
print("10! =", math.factorial(10))
print("6! =", math.factorial(6))

print("\nResultado final:")
print("Número de maneras de elegir los asistentes =", int(permutaciones))
```
