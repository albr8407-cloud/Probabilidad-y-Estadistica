```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

plt.figure(figsize=(6, 4))

media = 70
desv_std = 4
total_estudiantes = 500

peso1 = 68
peso2 = 72

prob1 = norm.cdf(peso2, media, desv_std)
prob2 = norm.cdf(peso1, media, desv_std)

probabilidad = prob1 - prob2
estudiantes = total_estudiantes * probabilidad

print("Probabilidad de que un estudiante pese entre 68 y 72 kg:", probabilidad)
print("Número de estudiantes que pesan entre 68 y 72 kg:", estudiantes)

x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso1) & (x <= peso2), color='skyblue', alpha=0.5)

plt.title('Probabilidad de que un estudiante pese entre 68 y 72 kg.\n μ=70, σ=4, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')

plt.grid(True)
plt.show()
```
